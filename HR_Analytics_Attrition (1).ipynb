```python
import pandas as pd

df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.head()
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns=['EmployeeCount','Over18','StandardHours'], inplace=True)
```

```python
df['Attrition_Flag'] = df['Attrition'].map({'Yes':1,'No':0})
```

```python
df['AnnualIncome'] = df['MonthlyIncome'] * 12
```

```python
df['Attrition'].value_counts(normalize=True) * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='Department', y='Attrition_Flag', data=df)
plt.title('Attrition Rate by Department')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='JobRole', y='Attrition_Flag', data=df)
plt.xticks(rotation=45)
plt.title('Attrition by Job Role')
plt.show()
```

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income vs Attrition')
plt.show()
```

```python
sns.barplot(x='WorkLifeBalance', y='Attrition_Flag', data=df)
plt.title('Work Life Balance vs Attrition')
plt.show()
```

```python
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Tenure vs Attrition')
plt.show()
```

### Key Insights

- The Sales department shows the highest attrition rate.
- Low-income employees have a higher probability of leaving.
- Poor work-life balance strongly impacts employee retention.
- Early-tenure employees are at higher risk of attrition.

### Business Recommendations
- Implement department-specific retention strategies for Sales.
- Improve compensation for high-risk employee segments.
- Enhance work-life balance through flexible policies.
- Strengthen onboarding and early employee engagement.
